# Excel- ja csv-muotoisten datojen avaaminen

```python
#Otan käyttöön pandas-ohjelmakirjaston:
import pandas as pd

#Avaan datan dataframeen, jonka nimeksi annan df1.
#Avaan datan data1.xlsx-tiedoston Data-taulukkovälilehdeltä.
#Vanhemmissa pandas-versioissa <b>sheet_name</b> sijasta pitää käyttää <b>sheetname</b>.
df1 = pd.read_excel('http://taanila.fi/data1.xlsx',sheet_name = 'Data')

#Näytän dataframen viisi ensimmäistä riviä:
df1.head()
```

```python
#Avaan datan dataframeen, jonka nimeksi annan df2.
#Avaan datan data1.csv-tiedostosta.
#data1.csv-tiedostossa on käytetty erottimena pilkun sijasta puolipistettä:
df2 = pd.read_csv('http://taanila.fi/data1.csv', delimiter = ';')

#Näytän dataframen kolme ensimmäistä riviä:
df2.head(3)
```
