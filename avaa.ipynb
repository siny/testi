### Excel- ja csv-muotoisten aineistojen avaaminen

```python
#pandas-kirjastossa on toiminnot aineiston avaamiseen
#Vakiintuneen tavan mukaisesti käytän tässä pandas sijasta lyhennettä pd

import pandas as pd
```

```python
#Avaan aineiston dataframeen, jonka nimeksi annan df1
#Avaan aineiston data1.xlsx-tiedoston Data-taulukkovälilehdeltä

df1=pd.read_excel('http://www.elisanet.fi/taanila/data1.xlsx',sheetname='Data')

#Tulostan aineiston sisällön

df1
```

```python
#Avaan paikalliselta levyltä csv-muotoisen aineiston
#Tässä aineistossa erotinmerkkinä on pilkun sijasta puolipiste

df2=pd.read_csv('C:/Users/taaak/data1.csv',delimiter=';')

#Tulostan aineiston 3 ensimmäistä riviä

df2.head(3)
```
