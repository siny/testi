<h2>csv-datan avaaminen pandas dataframeen</h2>

Pandas kirjaston <b>read_csv</b> avaa csv-muotoisen (comma separated value) datan.

Pelkällä tiedostonimellä avaat koodin kanssa samaan kansioon tallennetun csv-datan.

Jos tiedosto on muussa kansiossa tai netissä, niin lisää tiedostopolku tai nettiosoite (esim. <i>'C://Users/Aki/Documents/data1.csv'</i>).

Usein erottimena käytetään puolipistettä pilkun sijasta, jotta pilkkua voidaan käyttää desimaalierottimena. Tällöin tarvitset lisäparametreja <b>sep=';'</b> ja <b>decimal=','</b>.  

Datan alkuosaa voit avaamisen jälkeen katsoa komennolla <b>df.head()</b> ja loppuosaa komennolla <b>df.tail()</b>.

Lisätietoa read_csv käytöstä:
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html


```python
import pandas as pd
```

```python
df = pd.read_csv('http://taanila.fi/data1.csv', sep=';', decimal=',')
df.head()
```

```python
df.tail()
```

Jos erotinmerkki ja desimaalierotin eivät täsmää tiedoston kanssa tai jos tiedoston alussa on ylimääräistä selittävää tekstiä, niin read_csv antaa virheilmoituksen. Tällaisessa tapauksessa kannattaa nuuskia etukäteen tiedoston sisältöä muilla keinoin.

Voit avata tiedoston tekstieditoriin, esimerkiksi WordPadiin. Windowsin mukana tuleva WordPad on hyvä vaihtoehto, koska se avaa isotkin tiedostot (toisin kuin Notepad).

Jos tiedosto on omalla levyllä, niin nuuskiminen onnistuu Pythonista käsin esimerkiksi seuraavasti (esimerkkitiedoston kohdalla alussa on ylimääräinen rivi sep=;)

```python
tiedosto = open('elisa.csv','r')
print(tiedosto.readline())
print(tiedosto.readline())
print(tiedosto.readline())
tiedosto.close()
```

Jos tiedosto on netissä, niin nuuskiminen onnistuu esimerkiksi seuraavasti:

```python
import requests
tiedosto = requests.get('http://taanila.fi/elisa.csv')
print(tiedosto.text)
```

Edellisen perusteella avaaminen dataframeen onnistuu seuraavasti:

```python
df = pd.read_csv('http://taanila.fi/elisa.csv', sep=';', decimal=',', skiprows=1)
df.head()
```
