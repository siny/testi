```python
import pandas as pd

#Avaan Excel-datan dataframeen ja näytän 5 ensimmäistä riviä
df = pd.read_excel('http://www.elisanet.fi/taanila/data1.xlsx', sheet_name = 'Data')
df.head()
```

```python
#Minkä kokoinen data?
df.shape
```

```python
#Kuinka monta arvoa kussakin sarakkeessa?
df.count()
```

```python
#Näytän palkan mukaisessa järjestyksessä
df.sort_values('palkka')
```

```python
#Tallennan df_filtered1 dataframeksi ne rivit, joilla tyytyväisyys johtoon on pienempi kuin 3
df_filtered1 = df[df['tyytyväisyys johtoon'] < 3]
df_filtered1.head()
```

```python
#Tallennan df_filtered2 dataframeksi ne rivit, joilla palvelusvuosia on vähemmän kuin kaksi
#ja tyytyväisyys johtoon on 1 (erittän tyytymätön)
df_filtered2=df[(df['palvelusvuodet'] < 2) & (df['tyytyväisyys johtoon'] == 1)]
df_filtered2.head()
```

```python
#Lasken palkan päälle 5 % bonuksen
df['bonus'] = df['palkka'] * 0.05
df['yhteensä'] = df['palkka'] + df['bonus']
df[['palkka', 'bonus', 'yhteensä']].head()
```

```python
#Lisään vielä 3 prosentin lisäbonuksen alle 3000 palkoille
df['lisäbonus'] = (df['palkka'] < 3000) * 0.03 * df['palkka']
df[['palkka', 'bonus', 'yhteensä', 'lisäbonus']]
```

```python
#Lasken eri koulutuksen suorittaneiden lukumäärät lukumäärän mukaan järjestettynä
df['koulutus'].value_counts()
```

```python
#Lasken eri koulutuksen suorittaneiden lukumäärät koulutuksen mukaan järjestettynä
df['koulutus'].value_counts(sort = False)
```

```python
#Lasken tunnusluvut jokaiselle sarakkeelle
df.describe()
```

```python
#Lasken muuttujien väliset korrelaatiokertoimet
df.corr()
```

```python
#Lasken palkkakeskiarvon niille, joiden tyytyväisyys palkaan on 4 (tyytyväinen) tai 5 (erittäin tyytyväinen)
df['palkka'][df['tyytyväisyys palkkaan'] >= 4].mean()
```
