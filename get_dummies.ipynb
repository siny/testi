<h2>Dummy-muuttujat kategorisesta muuttujasta</h2>

Useimmat koneoppimisen mallit eivät pysty käyttämään kategorisia muuttujia selittävinä muuttujina. Voin kuitenkin muuntaa kategoriset muuttujat dummy-muuttujiksi, joita malleissa voi käyttää selittävinä muuttujina.

```python
import pandas as pd

# Avaan datan
df = pd.read_excel('http://taanila.fi/data1.xlsx')
df.head()
```

```python
# Teen koulutus-muuttujasta dummy-muuttujat pandas-kirjaston get_dummies -toiminnolla
dummies = pd.get_dummies(df['koulutus'], prefix = 'Koulutus')

# Yhdistän (concat) alkuperäisen datan ja koulutus-muuttujasta tehdyt dummy-muuttujat
df1 = pd.concat([df, dummies], axis = 1)
df1
```

```python
# Tässä sama kuin edellä, mutta samalla pudotan alkuperäisen koulutus-muuttujan pois datasta
df2 = pd.concat([df.drop('koulutus', axis = 1), dummies], axis = 1)
df2
```

```python
# Tässä tallennan "hölmösti" useamman tiedon samaan soluun
df3 = pd.DataFrame({
    'Nimi':['Mikko', 'Liisa', 'Veli', 'Tuukka', 'Eeva', 'Kaija'],
    'Arvosana':['4,2,3','3,1,2', '3,5,5', '2,4,4', '5,2,5', '3,3,3']})
df3
```

```python
# Edellisestä saan dummy-muuttujat seuraavasti
dummies = df3['Arvosana'].str.get_dummies(',')

# Yhdistän (concat) alkuperäisen datan ja Arvosana-muuttujasta tehdyt dummy-muuttujat
df4 = pd.concat([df3, dummies], axis = 1)
df4
```

```python
# Voin toki muuntaa "hölmösti" samaan soluun syötetyt arvosanat erillisiksi arvosanoiksi
df5 = pd.DataFrame(df3['Arvosana'].str.split(',', expand=True))
df5.columns=['matikka', 'fyssa', 'kemia']

#Yhdistän arvosanat df3 -dataframeen
df6 = pd.concat([df3, df5], axis=1)
df6
```
