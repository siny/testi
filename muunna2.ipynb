Seuraavassa tarkastelen transform()-toimintoa, logaritmimuunnosta ja standardointia (normittaminen).

```python
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_excel('http://taanila.fi/data1.xlsx')[['palkka', 'johto', 'työtov', 'työymp', 'palkkat', 'työteht']]
df.info()
```

```python
# Käännän mielipideasteikollisten muuttujien (1-5) asteikon päinvastaikseksi
# lambda tarkoittaa funktiota; tässä funktio suorittaa laskutoimituksen 6 - x, jokaiselle muuttujan arvolle
df[['johto2', 'työtov2', 'työymp2', 'palkkat2', 'työteht2']] = \
    df[['johto', 'työtov', 'työymp', 'palkkat', 'työteht']].transform(lambda x: 6 - x)
df.head()
```

```python
# Teen palkalle logaritmimuunnoksen kahdella vaihtoehtoisella tavalla
df['palkka_log1'] = df['palkka'].transform('log')
df['palkka_log2'] = np.log(df['palkka'])
df.head()
```

```python
# Kaksi vaihtoehtoista tapaa standardoida (normittaa)

# Keskiarvo
ka = df['palkka'].mean()

# Keskihajonta
# Huom. keskihajonnan kaavassa voi olla jakajana n tai n-1, tässä ddof-lisäparametri määrää jakajaksi n
kh = df['palkka'].std(ddof=0)

# Tapa 1
df['palkka_norm1'] = (df['palkka'] - ka) / kh

# Tapa 2
df['palkka_norm2'] = df['palkka'].apply(lambda x: (x - ka) / kh)
```
