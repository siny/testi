```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel('http://www.elisanet.fi/taanila/opintomenestys.xlsx')
df.head()
```

```python
#Tunnusluvut
df.describe()
```

```python
#Korrelaatiot
df.corr()
```

```python
#Kielillä oli suurin korrelaatio opintomenestyksen kanssa
plt.scatter(df['kielet'], df['opintomenestys'])
plt.xlabel('kielet')
plt.ylabel('opintomenestys')
plt.savefig('kuva1.png') #Kuvan voi tallentaa png, pdf, ps, eps and svg -muodoissa
plt.show()
```
