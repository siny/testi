Esimerkkejä aikasarjoista ja Quandl-palvelun käytöstä

```python
import quandl
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
quandl.ApiConfig.api_key = 'xxx' #tähän tarvitaan oma quandl-salasana
```

```python
oil = quandl.get("EIA/PET_RWTC_D")
oil.head()
```

```python
oil.columns=['WTI Oil spot price']
oil.plot()
```

```python
oil['2016':].plot() #vuoden 2016 alusta
```

```python
al=quandl.get("LME/PR_AL") #alumiinin hintoja
al.head()
```

```python
cu=quandl.get("LME/PR_CU") #kuparin hintoja
cu.head()
```

```python
al_cu=pd.concat([al['Cash Buyer'], cu['Cash Buyer']], axis=1) #Yhdistetään alumiinin ja kuparin Cash Buyer samaan aineistoon
al_cu.columns=['Alumiini', 'Kupari']
al_cu.plot()
```

```python
#Figure-olio voi sisältää yhden tai useampia axes-olioita (kuvioita)
#Figure ja axes -oliot luodaan plt.subplots-funktiolla
fig, ax1 = plt.subplots(figsize=(10,6))

color = 'royalblue'
ax1.set_xlabel('Vuosi')
ax1.set_ylabel('Alumiini', color=color)
ax1.plot(al_cu['Alumiini'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'darkorange'
ax2.set_ylabel('Kupari', color=color)
ax2.plot(al_cu['Kupari'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
al_cu=al_cu['2017':] #vuoden 2017 alusta
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'royalblue'
ax1.set_xlabel('Aika')
ax1.set_ylabel('Alumiini', color=color)
ax1.plot(al_cu['Alumiini'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'darkorange'
ax2.set_ylabel('Kupari', color=color)
ax2.plot(al_cu['Kupari'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
microsoft=quandl.get("EOD/MSFT")
microsoft.head()
```

```python
microsoft['Close'].plot()
microsoft['Close'].rolling(250).mean().plot() #250 päivän liukuva keskiarvo
```

```python
microsoft['Close']['2015':].plot() #Vuoden 2015 alusta
microsoft['Close']['2015':].rolling(250).mean().plot()
```

```python
microsoft['Close']['2015':].resample('BM').mean().plot() #kuukausikeskiarvot
```

```python
microsoft['Close']['2015':].asfreq('BM').plot() #kuukauden viimeisen päivän hinnat
```

```python
microsoft['Close']['2018'].plot()
microsoft['Close']['2018'].ewm(alpha=0.3).mean().plot() #Eksponentiaalinen tasoitus
```

```python
apple=quandl.get("EOD/AAPL")
mi_ap=pd.concat([microsoft['Close']['2018'], apple['Close']['2018']], axis=1) #Yhdistetään Microsoft ja Apple samaan aineistoon
mi_ap.columns=['Microsoft', 'Apple']
mi_ap.head()
```

```python
mi_ap.plot()
```

```python
mi_ap['mic muutos']=mi_ap['Microsoft'].pct_change() #Prosenttimuutokset
mi_ap['app muutos']=mi_ap['Apple'].pct_change()
mi_ap
```

```python
corr=mi_ap['mic muutos'].rolling(125, min_periods=30).corr(mi_ap['app muutos']) #125 päivän liukuva korrelaatio
corr.plot()
```

```python
walt_disney=quandl.get("EOD/DIS")
```

```python
mi_ap_wa=pd.concat([microsoft['Close'], apple['Close'], walt_disney['Close']], axis=1) #Yhdistetään kolmen osakkeen tiedot samaan aineistoon
mi_ap_wa.columns=['Microsoft', 'Apple', 'Walt Disney']
mi_ap_wa.plot()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(mi_ap_wa['2018'])
ax1.legend(mi_ap_wa.columns)
```

```python
writer = pd.ExcelWriter('mi_ap_wp.xlsx', engine='xlsxwriter')
mi_ap_wa.to_excel(writer, sheet_name = 'aikasarjat') #Aikasarjat Exceliin
writer.save()
```
