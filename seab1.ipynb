```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")

df = pd.read_excel('http://www.elisanet.fi/taanila/opintomenestys.xlsx')
df.head()
```

```python
sns.pairplot(df[['verbaalinen','looginen','kielet','matematiikka','opintomenestys']], kind='reg')
```

```python
sns.pairplot(df, hue='sukupuoli')
```

```python
sns.jointplot(x='kielet', y='opintomenestys', data=df, kind='reg')
```

```python
from scipy.stats import spearmanr
sns.jointplot(x='matematiikka', y='opintomenestys', data=df, kind='reg', stat_func=spearmanr)
```

```python
#distplot estimoi havaintojen taustalla olevaa todennäköisyysjakaumaa KDE-mentelmällä (kernel density estimation)
sns.distplot(a=df['verbaalinen'])
```

```python
sns.boxplot(x='matematiikka', y='opintomenestys', data=df)
```

```python
sns.boxplot(x='matematiikka', y='opintomenestys', hue='sukupuoli', data=df)
```
