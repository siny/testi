<h2>Ohjelmakirjastot</h2>

Aluksi tuon (<b>import</b>) tarvittavat kirjastot. Data-analytiikassa tarvitaan aina <b>pandas</b>-kirjastoa. Pandas sisältää <b>dataframe</b>-tietorakenteen, jota tarvitaan taulukkomuotoiselle datalle. Vakiintuneen tavan mukaan käytän pandas-kirjastosta lyhennettä <b>pd</b>.

```python
import pandas as pd
```

<h2>Datan avaaminen</h2>

Excel-muotoisen datan voin lukea pandas-kirjaston <b>read_excel()</b>-funktiolla. <b>head()</b>-funktio tulostaa datan viisi ensimmäistä riviä, <b>tail()</b>-funktio tulostaa viisi viimeistä riviä ja <b>shape</b>-ominaisuus antaa datan rivien ja sarakkeiden lukumäärän.

```python
df = pd.read_excel('http://taanila.fi/data1.xlsx')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

<h2>Sarakkeet</h2>

Sarakkeiden (muuttujien) nimilistan saan tulostamalla <b>columns</b>-ominaisuuden. Voin vaihtaa muuttujien nimiä sijoittamalla muutetun nimilistan columns-ominaisuuden arvoksi. Muuttujien nimiä voin vaihtaa myös <b>rename()</b>-funktiolla.

Muuttujien arvojen lukumäärät selviävät <b>count()</b>-funktiolla.

```python
df.columns
```

```python
#sijoitan muutetun nimilistan muuttujien nimiksi
df.columns = ['nro', 'sukupuoli', 'ikä', 'perhesuhde', 'koulutus', 'palveluv', 'palkka',
       'johto', 'työtov', 'työymp', 'palkkat', 'työteht', 'työterv', 'lomaosa',
       'kuntosa', 'hieroja']
df.columns
```

```python
#vaihdan muuttujien nimiä rename()-funktiolla
df.rename(columns = {'sukupuoli': 'sukup', 'perhesuhde': 'perhe'}, inplace = True)
df.head()
```

```python
df.count()
```

```python
#puuttuvien havaintojen lukumäärät muuttujittain
df.isnull().sum()
```

<h2>Ainutkertaiset arvot</h2>

Muuttujan sisältämät ainutkertaiset arvot selviävät <b>unique()</b>-funktiolla. Seuraavassa käyn läpi kaikki muuttujat <b>for</b>-silmukalla.

```python
for var in df:
    print(var, pd.unique(df[var]))
```

<h2>Uusien muuttujien laskeminen</h2>

<b>replace()</b>-funktiolla korvaan muuttujan arvoja uusilla arvoilla.
<b>cut()</b>-funktiolla luokittelen määrällisen muuttujan arvot luokkiin.
<b>mean()</b> ja <b>sum()</b> -funktioilla voin laskea keskiarvo- ja summamuuttujia. Lisäparametri <b>axis=1</b> tarkoittaa laskemista vierekkäin samalla rivillä sijaitsevista luvuista.

```python
#uusi muuttuja 'sukup_teksti', jossa sukupuolet tekstiarvoina
df['sukup_teksti'] = df['sukup'].replace({1 : 'Mies', 2 : 'Nainen'})
df.head(6)
```

```python
#uusi muuttuja 'ikäluokka', jossa iät luokiteltuina ikäluokkiin
bins = [18, 28, 38, 48, 58, 68]
df['ikäluokka'] = pd.cut(df['ikä'], bins = bins)
df.head()
```

```python
#lasken vastaajan "kokonaistyytyväisyyden" keskiarvona
df['tyytyväisyys'] = df[['johto','työtov','työymp','palkkat','työteht']].mean(axis=1)

df[['sukup','ikä','tyytyväisyys']].head()
```

```python
#lasken käytettyjen etuisuuksien lukumäärän.
df['käyttö'] = df[['työterv', 'lomaosa', 'kuntosa', 'hieroja']].count(axis=1)
df.head(6)
```

<h2>Poimintaa ja suodatuksia</h2>

Voin muodostaa monimutkaisiakin poimintoja <b>&</b>- ja <b>|</b>-operaattoreita käyttäen.

<b>notnull()</b>- ja <b>isnull()</b>-funktioilla voin selvittää ei-tyhjät ja tyhjät.

<b>sort_values()</b>-funktio järjestää halutun muuttujan suhteen.

<b>drop()</b>-funktio pudottaa halutut muuttujat pois datasta.

```python
#huomaa, että riviä 6 ei oteta mukaan
df[4:6]
```

```python
#ne, joiden palkka on suurempi kuin 4000
df[df['palkka']>4000]
```

```python
#miehet, joiden tyytyväisyys palkaan on 1 (erittäin tyytymätön) tai 2 (tyytymätön)
df[(df['sukup']==2) & (df['palkkat']<3)]
```

```python
#ne, jotka ovat käyttäneet sekä työterveyshuoltoa että lomaosaketta
df[(df['työterv'].notnull()) & (df['lomaosa'].notnull())]
```

```python
#ne, jotka eivät ole käyttäneet mitään tarjotuista etuisuuksista
df[df['käyttö']==0]
```

```python
#kokonaistyytyväisyys niille, jotka ovat käyttäneet vähintään kolmea etuisuutta
df[['tyytyväisyys','käyttö']][df['käyttö']>=3]
```

```python
#Palkat ja palkkatyytyväisyydet palkan mukaan järjestettynä
df[['palkka','palkkat']].sort_values(by='palkka').head()
```

```python
#sama kuin edellä, mutta palkat laskevassa järjestyksessä
df[['palkka','palkkat']].sort_values(by='palkka', ascending=False).head()
```

```python
#data ilman muuttujia nro, ikä, palveluv ja palkka
df.drop(['nro','ikä','palveluv','palkka'], axis=1).head()
```
