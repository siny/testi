<h2>yahoo finance</h2>

Yahoo finance https://finance.yahoo.com/ tarjoaa tickerin avulla haettavia aikasarjoja.

Tietojen hakemiseen tarvitaan pandas_datareader, joka ei ole Anacondaan valmiiksi asennettuna.

```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
%matplotlib inline
```

```python
# Haetaan historiatiedot Nokian osakkeesta vuoden 2020 alusta (NYSE)
aapl_2020 = web.DataReader('NOK', start='2020-1-1', data_source='yahoo')
aapl_2020
```

```python
# Haetaan Applen osakkeen hintahistoria vuoden 2015 alusta
aapl = web.DataReader('AAPL', start='2015-1-1', data_source='yahoo')['Close']

aapl.plot(title='Apple - osakkeen hintakehitys')
```

```python
# Haetaan kullan hintakehitys vuoden 2015 alusta
gold = web.DataReader('GC=F', start='2015-1-1', data_source='yahoo')['Close']

gold.plot(title='Kullan hintakehitys')
```

```python
# Kahden arvoakselin kaavio Applen ja kullan hintahistoriasta

ax1 = aapl.plot(color = 'C0')
ax1.set_ylabel('Apple', color='C0', fontsize=14)

ax2 = ax1.twinx()
gold.plot(ax = ax2, color='C1')
ax2.set_ylabel('Gold', color='C1', fontsize=14)
```
